```python
import pandas as pd

gender_submission =test  = pd.read_csv('train.csv')
test  = pd.read_csv('test.csv')
train = pd.read_csv('gender_submission.csv')
gender_submission.head(), test.head(), train.head()
train.info()

train.describe()
train = pd.read_csv('train.csv')


# Strip any whitespace from column names
train.columns = train.columns.str.strip()

# Handle missing values
train['Age'] = train['Age'].fillna(train['Age'].median())
train['CabinKnown'] = train['Cabin'].notna().astype(int)
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])

# Verify there are no missing values
print(train.info())


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# 1. Survival Rate by Passenger Class
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=train, palette='inferno')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()

# 2. Survival Rate by Sex
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=train, palette='inferno')
plt.title('Survival Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()



```

```python
# 3. Survival Rate by Age
plt.figure(figsize=(10, 6))
sns.histplot(data=train, x='Age', hue='Survived', multiple='stack', palette='inferno', bins=30)
plt.title('Survival Count by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 4. Survival Rate by SibSp
plt.figure(figsize=(10, 6))
sns.countplot(x='SibSp', hue='Survived', data=train, palette='inferno')
plt.title('Survival Count by Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
# 5. Survival Rate by Parch
plt.figure(figsize=(10, 6))
sns.countplot(x='Parch', hue='Survived', data=train, palette='inferno')
plt.title('Survival Count by Number of Parents/Children Aboard')
plt.xlabel('Number of Parents/Children Aboard')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()

# 6. Survival Rate by Fare
plt.figure(figsize=(10, 6))
sns.histplot(data=train, x='Fare', hue='Survived', multiple='stack', palette='inferno', bins=30)
plt.title('Survival Count by Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

```

```python
# 7. Survival Rate by Embarked
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', hue='Survived', data=train, palette='inferno')
plt.title('Survival Count by Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()

# 8. Survival Rate by CabinKnown
plt.figure(figsize=(10, 6))
sns.countplot(x='CabinKnown', hue='Survived', data=train, palette='inferno')
plt.title('Survival Count by Cabin Known')
plt.xlabel('Cabin Known')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```
